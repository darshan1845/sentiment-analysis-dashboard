```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

import nltk 
from textblob import TextBlob

print("All libraries imported successfully")
```

```python
nltk.download('punkt')
```

```python
text = "I really love this product. It is amazing!"

blob = TextBlob(text)
print("Polarity:", blob.sentiment.polarity)
print("Subjectivity:", blob.sentiment.subjectivity)
```

```python
def get_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
print(get_sentiment("I love this movie"))
print(get_sentiment("This is the worst product ever"))
print(get_sentiment("It is a pen"))
```

```python
data = {"text": 
        ["I love this phone, the battery life is amazing",
        "This is the worst service I have ever experienced",
        "The product is okay, not too good not too bad",
        "Very happy with the customer support",
        "I am disappointed with the quality"]}

df = pd.DataFrame(data)
df
```

```python
df["sentiment"] = df["text"].apply(get_sentiment)
df
```

```python
sentiment_counts = df["sentiment"].value_counts()
sentiment_counts
```

```python
plt.figure()
sentiment_counts.plot(kind='bar')
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sentiment_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sentiment Percentage")
plt.ylabel("")
plt.show()
```
